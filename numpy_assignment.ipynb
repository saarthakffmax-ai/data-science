Objective: Understand NumPy array creation and manipulation.

```
import numpy as np

arr = np.arange(1, 21)
print(arr)
print('Sum:', arr.sum())
print('Mean:', arr.mean())
print('Median:', np.median(arr))
print('Std Dev:', arr.std())
print('Indices > 10:', np.where(arr > 10))
```

```
arr2d = np.arange(1, 17).reshape(4, 4)
print(arr2d)
print('Transpose:\n', arr2d.T)
print('Row sums:', arr2d.sum(axis=1))
print('Column sums:', arr2d.sum(axis=0))
```

```
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))
print(a)
print(b)
print('Add:\n', a+b)
print('Sub:\n', a-b)
print('Mul:\n', a*b)
print('Dot:\n', a.dot(b))
```
